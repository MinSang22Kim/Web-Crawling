```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd

data=[]
response = requests.get(f"https://www.dangyoung.com/")
html = response.text
soup = BeautifulSoup(html, 'html.parser')
items = soup.select(".w-100")

for item in items:
    company = item.select_one(".HomeComponent_brandName__ZlB0w").text
    name = item.select_one(".HomeComponent_name__GJ8ZY").text
    detail_1 = item.select_one(".HomeComponent_nutrition__W_vjw").text
    price = item.select_one(".HomeComponent_price__2Rkat").text.replace('원','')
    data.append([company, name, detail_1, price])

df = pd.DataFrame(data, columns=['회사', '이름', '설명', '가격'])
df
```
